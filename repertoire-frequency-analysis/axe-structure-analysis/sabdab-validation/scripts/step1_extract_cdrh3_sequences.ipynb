```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
dirpath = '/Users/ssolieva/Desktop/github_repo/Q23_paper/analysis-of-HH/sabdab-validation/sabdab_antibodies/20240804_0567926/imgt/'

pdb_paths_file = []
with open(f'{dirpath}pdb_paths.txt') as f:
    pdb_paths_file.append([line.rstrip() for line in f])
pdb_paths = pdb_paths_file[0]
print(len(pdb_paths), pdb_paths[0])
```

```python
HC_seq = []
pdb_names =[]
for i in range(len(pdb_paths)):
    pdb_names.append(f'{pdb_paths[i][-8:-4]}')
    pdb_structure=md.load(pdb_paths[i])
    #print(pdb_structure, pdb_paths[i])
    HC_seq_atoms = pdb_structure.topology.select('protein and chainid 0 and name CA')
    HC_ResSeq =[]
    for j in range(len(HC_seq_atoms)):
        HC_ResSeq.append(pdb_structure.topology.atom(HC_seq_atoms[j]).residue.code)
    HC_seq.append(''.join(HC_ResSeq))
```

```python
# Get CDRH3 sequences so RMSD can be calculated for only the cdrh3 regions. 
# step 1: search for cdrh3 using the first few resis of CDRH3 sequences 
start_resis = ['CAR', 'CAK', 'CVK', 'CAA', 'CAT', 'CAP', 'CAQ', 'CAS', 'CAF', 'CAI', 'CAG', 'CAL', 
               'CTT', 'CTR', 'CVR', 'CVQ', 'CHR', 'CTW', 'CAN', 'CGR', 'CVS', 'CNL', 'CCT', 'CSR', 
               'CSQ','CSG','CTK', 'CAH','CKT','CVI','CTI','CRT','CMT','CTG','CLT','CST','CGA','CRV','CVT','CAY','CTS','CSK','CIR'
              ]#, 'CSG']#, 'CSG']#, 'YYCA']
cdrh3_seq = []
found_indices = [] 
pdb_name_cdrh3 = []
pdb_path_cdrh3 = []

for i in range(len(HC_seq)):
    for j in range(len(start_resis)):
        if HC_seq[i][80:].find(start_resis[j]) > 0:
            index=HC_seq[i][80:].find(start_resis[j])+81 # don't include 'c'
            cdrh3_seq.append(HC_seq[i][index:])#[HC_seq[i].find(start_resis[j])+1+80:])
            found_indices.append(i)
            pdb_name_cdrh3.append(pdb_names[i])
            pdb_path_cdrh3.append(pdb_paths[i])
            break
            
for i in range(len(HC_seq)):
    if i not in found_indices:
        print('\t seq. not found in search: ', HC_seq[i][80:], i, pdb_names[i])

if len(cdrh3_seq) != len(found_indices) != len(np.unique(found_indices)):
    print('ERROR: mismatched # of sequences')
```

```python

# step 2: get rid of the end of the cdrh3 using last few resis of cdrh3 sequences. 
end_resis = ['WGQG', 'WGTG', 'WGPG', 'WGKG', 'WGRG', 'WGAG', 'WGLG', 'WGHG', 'WGRK', 'WGCG', 'WSQG', 'WGEG','WAKG', 'WGNG','WGSG']
cdrh3_seq_updated = []
cdrh3_seq_updated_not_found = []
cdrh3_seq_updated_pdb_name = []
pdb_path_cdrh3_updated =[]
for i in range(len(cdrh3_seq)):
    for j in range(len(end_resis)):
        if cdrh3_seq[i].find(end_resis[j]) > 0:
            cdrh3_seq_updated.append(cdrh3_seq[i][:cdrh3_seq[i].find(end_resis[j])])
            cdrh3_seq_updated_pdb_name.append(pdb_name_cdrh3[i])
            pdb_path_cdrh3_updated.append(pdb_path_cdrh3[i])
            break
            
if len(cdrh3_seq) != len(cdrh3_seq_updated):
    print('ERROR: mismatched # of sequences')
    print(len(cdrh3_seq),len(pdb_path_cdrh3_updated))
```

```python
# filter out cdrh3 seqs < 10 and > 40:

cdrh3_seq_updated_filtered = []
pdb_path_cdrh3_updated_filtered = []
cdrh3_seq_updated_pdb_name_filtered = []

for i in range(len(cdrh3_seq_updated)):
    if len(cdrh3_seq_updated[i]) > 9 and len(cdrh3_seq_updated[i]) < 40:
        cdrh3_seq_updated_filtered.append(cdrh3_seq_updated[i])
        pdb_path_cdrh3_updated_filtered.append(pdb_path_cdrh3_updated[i])
        cdrh3_seq_updated_pdb_name_filtered.append(cdrh3_seq_updated_pdb_name[i])

print(len(cdrh3_seq_updated_pdb_name_filtered))
```
