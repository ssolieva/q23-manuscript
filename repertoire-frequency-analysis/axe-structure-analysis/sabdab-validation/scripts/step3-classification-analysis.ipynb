```python
beta_sheet_distance_cutoff = 20 #20A
classification_list=[]

for i in range(120,200):#len(cdrh3_seqs)):
    try:
        #print(i)
        name = ' '
        pdb_path = pdb_paths[i]
        cdrh3_sequence = cdrh3_seqs[i]
        phi_angles, psi_angles, dssp_values, hbonds3, hbonds4, distances = run_all_commands(name, pdb_path, cdrh3_sequence)
        features = [cdrh3_sequence, phi_angles, psi_angles, dssp_values, hbonds3, hbonds4, distances]
        csv_file = generate_dataframe(features) # not saving csv for space.
        turns_index,   beta_index    = classify_and_combine(csv_file)
        hbonds = classify_hbonds(csv_file['hbondi_i3'], csv_file['hbondi_i4'])
        combined_TB = combine_all_dist(beta_index, turns_index, hbonds, distances,beta_sheet_distance_cutoff)
        classification_of_hh = classify_turn_beta_distance_v2(csv_file, name, combined_TB, beta_sheet_distance_cutoff)
        if classification_of_hh == 1:
            print(f'{i},yesHH,{pdb_path},{cdrh3_sequence}')
            classification_list.append(f'yesHH,{pdb_path},{cdrh3_sequence}')
        if classification_of_hh == 0:
            print(f'noHH,{pdb_path},{cdrh3_sequence}')
            classification_list.append(f'noHH,{pdb_path},{cdrh3_sequence}')
    except:
        pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('classification_results.txt') as f:
    data_full = [line.rstrip() for line in f]
```

```python
pdb_paths = []
classification = []
cdrh3_lengths=[]
for i in range(len(data_full)):
    if len(data_full[i].split(',')) > 1:
        if data_full[i].split(',')[1] not in pdb_paths:
            pdb_paths.append(data_full[i].split(',')[1])
            classification.append(data_full[i].split(',')[0])
            cdrh3_lengths.append(len((data_full[i].split(',')[2])))
            if len((data_full[i].split(',')[2])) > 30:
                print(len((data_full[i].split(',')[2])), data_full[i].split(',')[1])
```

```python
count_yes = []
count_no = []
cdrh3_lengths_yes = []
for i in range(len(classification)):
    if classification[i] == 'yesHH':
        count_yes.append(1)
        cdrh3_lengths_yes.append(cdrh3_lengths[i])
        print(pdb_paths[i], )
    if classification[i] == 'noHH':
        count_no.append(1)
```

```python
cdrh3_lengths_yes
```

```python
plt.figure(figsize=[6,3])
plt.hist(cdrh3_lengths, edgecolor='k', bins=10, color='skyblue', linewidth=5)
plt.xticks(fontsize=16, weight='bold')
plt.yticks(fontsize=16, weight='bold')
plt.ylabel('# of structures', fontsize=16, weight='bold')
plt.xlabel('CDRH3 length', fontsize=16, weight='bold')
plt.tight_layout()
plt.savefig('cdrh3_dist.png',dpi=300,transparent=True)
plt.show()
```

```python
yesHH = len(count_yes)
noHH  = len(count_no)

fig, ax = plt.subplots(figsize=[4,4])
ax.pie([yesHH, noHH], 
       wedgeprops = {"edgecolor" : "black", 'linewidth': 1}, 
       autopct='%1.1f%%', colors=['#48bcbc','#96a0b3'])  #['#5596ce','#f29843','#e9c64e']
plt.legend([f'{yesHH} hammerheads', f'{noHH} non-hammerheads'], 
           bbox_to_anchor = (1.2, 0.1), fontsize=14)
plt.title(f' Counts', weight='bold', y=0.95, fontsize=16)
plt.tight_layout()
#plt.savefig(f'plots/{which_system}_pie_chart.pdf', dpi=300, transparent=True, bbox_inches="tight")
plt.show()
```
